Problem 12 - Highly divisible triangular number.
====

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be $1 + 2 + 3 + 4 + 5 + 6 + 7 = 28$. The first ten terms would be:

$1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...$

Let us list the factors of the first seven triangle numbers:

     1: 1
     3: 1,3
     6: 1,2,3,6
    10: 1,2,5,10
    15: 1,3,5,15
    21: 1,3,7,21
    28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

First we need to generate triangle numbers
---

We'll write a generator that works just like range(), but returns triangle number.

Note how we use "*args" here - it allows us to allow the function to take a TBD number of inputs, which we then pass on to the range() generator we use inside the function.

```python
def triangle(*args):
    for n in range(*args):
        yield n*(n+1)/2
```

```python
# Let's test our generator by generating the first 7 triangle numbers
triangle(7)
```

```python
def has_n_divisors(x, n):
    divisors = 2  # Count 1 and X as divisors automatically
    for i in range(2,x):
        if (x % i) == 0:
            divisors += 1
            if divisors >= n:
                return True
    return False
```

```python
target = 200
found = False
for i in triangle(1,5000):
    if has_n_divisors(i, target):
        print "Solution found with triangle number", i
        found = True
        break
        
if not found:
    print "Searched up to", i, "but solution not found"
```
